1) Write a function named `is_vowel`. It should accept a string as input and use a regular expression to determine if the passed string is a vowel. While not explicity mentioned in the lesson, you can treat the result of `re.search` as a boolean value that indicates whether or not the regular expression matches the given string.

```python
def is_vowel(string):
    vowel_re = r'^[aeiou]$'
    return re.match(vowel_re ,string.lower())is not None
```

```python
for i in 'aBcdE,~':
    print(f">>>> is_vowel('{i}')") 
    print(is_vowel(i))
```

2) Write a function named `is_valid_username` that accepts a string as input. A valid username starts with a lowercase letter, and only consists of lowercase letters, numbers, or the `_` character. It should also be no longer than 32 characters. The function should return either `True` or `False` depending on whether the passed string is a valid username.

```python
def is_valid_username(string):
    valid_username_re = r'^[a-z][a-z0-9_]{,31}$'
    return re.match(valid_username_re, string) is not None
```

```python
for i in ['aaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaa', 'codeup','code_up','Codeup','codeup123','1codeup', 'cOdEuP', 'x']:
    print (f">>>> is_valid_username('{i}')")
    print(is_valid_username(i))
```

3) Write a regular expression to capture phone numbers. It should match all of the following:

```python
numbers = pd.Series(['(210) 867 5309',
           '+1 210.867.5309',
           '867-5309',
           '210-867-5309'])

numbers.str.extract(r'''
^
(?P<country_code>\+\d)?
\D*?
(?P<area_code>\d{3})?
\D*?
(?P<prefix>\d{3})
\D*?
(?P<line>\d{4})
\D*
$
''', re.VERBOSE)
```

4) Use regular expressions to convert the dates below to the standardized year-month-day format.



```python
dates = ['02/04/19', '02/05/19',
          '02/06/19', '02/07/19',
          '02/08/19', '02/09/19',
          '02/10/19']
for i in dates:
    print(re.sub(r'(\d{2})/(\d{2})/(\d{2})', r'20\3-\1-\2',i))
```

6) You can find a list of words on your mac at /usr/share/dict/words. Use this file to answer the following questions:



```python
file= open('/usr/share/dict/words')
words = pd.read_csv(file, names = ['word'])
words.head()
```

- How many words have at least 3 vowels?

```python
words.word.apply(lambda x: re.match(r'\w*[aeiou]\w*[aeiou]\w*[aeiou]\w*', str(x).lower()) is not None).sum()
```

- How many words have at least 3 vowels in a row?

```python
words.word.apply(lambda x: re.match(r'.*[aeiou]{3}.*', str(x).lower()) is not None).sum()
```

- How many words have at least 4 consonants in a row?

```python
words.word.apply(lambda x: re.match(r'.*[^aeiou]{4}.*', str(x).lower()) is not None).sum()
```

- How many words start and end with the same letter?

```python
words.word.apply(lambda x: re.match(r'^(\w)\w*\1$', str(x).lower()) is not None).sum()
```

- How many words start and end with a vowel?

```python
words.word.apply(lambda x: re.match(r'^[aeiou].*[aeiou]$', str(x).lower()) is not None).sum()
```

- How many words contain the same letter 3 times in a row?

```python
words.word.apply(lambda x: re.match(r'\w*(\w)\1{2}\w*', str(x).lower()) is not None).sum()
```

- What other interesting patterns in words can you find?
