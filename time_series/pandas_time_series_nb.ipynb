###### San Francisco

```python
from vega_datasets import data
df = data.sf_temps()
```

```python
df.date = pd.to_datetime(df.date)
df.set_index('date', inplace=True)
```

1. Resample by the day and take the average temperature. Visualize the average temperature over time.


```python
df.resample('d').mean().plot()
plt.title('San Francisco-Average Daily Temp')
plt.ylabel('Temperature (f)')
```

2. Write the code necessary to visualize the minimum temperature over time.
3. Write the code necessary to visualize the maximum temperature over time.


```python
df.temp.resample('d').agg(['min','max','mean']).plot()
```

4. Which month is the coldest, on average?


```python
monthly_avg = df.resample('m').mean()
min_temp = monthly_avg.min()[0]
coldest_month = monthly_avg[monthly_avg.temp == min_temp].index.month_name()[0]
print(f'The coldest average month is {coldest_month}')
```

5. Which month has the highest average temperature?


```python
monthly_avg = df.resample('m').mean()
max_temp = monthly_avg.max()[0]
hottest_month = monthly_avg[monthly_avg.temp == max_temp].index.month_name()[0]
print(f'The hottest average month is {hottest_month}')
```

6. Resample by the day and calculate the min and max temp for the day. Use this resampled dataframe to calculate the change in temperature for the day. Which month has the highest daily temperature variability?


```python
daily_temp_range = df.resample('d').agg(['min', 'max'])
daily_temp_range['range'] = daily_temp_range.temp['max'] - daily_temp_range.temp['min']
monthly_avg = daily_temp_range.range.resample('m').mean()
max_range = monthly_avg.max()
variable_month = monthly_avg[monthly_avg == max_range].index.month_name()[0]
print(f'The month with the most variable temperatures is {variable_month}')
```

###### Seattle

```python
df = data.seattle_weather()
df.date = pd.to_datetime(df.date)
df.set_index('date', inplace = True)
df.head()
```

1. Which year and month combination has the highest amount of precipitation?


```python
total_precip = df.precipitation.resample('m').sum()
max_precip = total_precip.max()
wet_month = total_precip[total_precip == max_precip].index.strftime('%b,%Y')[0]
print(f'The month with the most precipitation is {wet_month}')
```

2. Visualize the amount of monthly precipitation over time.


```python
total_precip.plot()
plt.ylabel('Total Precipitation')
plt.xlabel('Month')
```

3. Visualize the amount of wind over time. Choose a time interval you think is appropriate.


```python
df.wind.resample('m').mean().plot()
```

4. Which year-month combination is the windiest?


```python
avg_wind = df.precipitation.resample('m').mean()
max_wind = avg_wind.max()
windy_month = avg_wind[avg_wind == max_wind].index.strftime('%b,%Y')[0]
print(f'The month with the highest average wind was {windy_month}')
```

5. What's the sunniest year? 

```python
sunny_days = df[df.weather == 'sun'].weather.resample('y').count() 
sunniest = sunny_days.max()
sunny_year = sunny_days[sunny_days == sunniest].index.year[0]
print(f'The sunniest year was {sunny_year}')
```

7. In which month does it rain the most?


```python
rainy_days = (df.weather == 'rain') | (df.weather == 'drizzle')
rainy_months = rainy_days.groupby(rainy_days.index.month_name()).mean()
rainniest = rainy_months.max()
wet_month = rainy_months[rainy_months == rainniest].index[0]
print(f'The ranniest month is {wet_month}')
```

###### flights

```python
df = data.flights_20k()
df.date = pd.to_datetime(df.date)
df.set_index('date', inplace = True)
df.head()
```

1. Convert any negative delays to 0.

```python
df.delay = df.delay.apply(lambda x: 0 if x < 0 else x)
(df.delay >= 0).sum()
```

2. Which hour of the day has the highest average delay?

```python
delays = df.delay.groupby(df.index.hour).mean()
highest_delay_hour = delays[delays == delays.max()].index[0]
print(f'The hour with the highest average delay time is {highest_delay_hour} o\'clock')
```

3. Does the day of the week make a difference in the delay amount?

```python
delays_by_day = (df.delay > 0).groupby((df.delay > 0).index.weekday_name).agg(['mean','count'])
delays_by_day['mean'].plot(kind = 'bar')
```

4. Does the month make a difference in the delay amount?

```python
delays_by_month = (df.delay > 0).groupby((df.delay > 0).index.month_name()).mean()
delays_by_month.plot(kind = 'bar')
```

###### Iowa Electricity

```python
df = data.iowa_electricity()
df.year = pd.to_datetime(df.year).dt.year
df.set_index('year', inplace = True)
df.head()
```

```python
totals = df.groupby('year').sum().rename(columns = {'net_generation':'totals'})
tall_df = df.join(totals)
tall_df['pct_year'] = tall_df.net_generation / tall_df.totals
tall_df.head()
```

```python
wide_df = df.pivot( columns= 'source', values = 'net_generation')
```

```python
wide_df.plot.area()
plt.title('Total Energy', )
plt.show()
```

```python
tall_df.drop(columns='net_generation').pivot(columns='source', values='pct_year').plot.area()
plt.title('Energy Breakdown')
```
