```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import TimeSeriesSplit
```

```python
solar = pd.read_csv('solar.csv')
solar.info()
```

```python
solar.rename(columns= 
             {'Date/Time':'date', 'Energy Produced (Wh)': 'output'}, 
             inplace= True
            )
solar.drop(index = 387, inplace=True)
solar.date = solar.date.str.split().apply(pd.Series)[0]
```

```python
solar.date = pd.to_datetime(solar.date)
df = solar.set_index('date')
df.head()
```

```python
df = df['2018-11-7':]
df.plot()
```

```python
df.resample('ms').mean().plot()
plt.show()
```

```python
df.resample('m').mean().plot()
plt.show()
```

```python
df.resample('q').mean().plot()
```
